```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "monospace"

from sklearn.metrics import \
    auc, \
    roc_auc_score, \
    accuracy_score, \
    jaccard_score, \
    f1_score, \
    precision_score, \
    recall_score, \
    precision_recall_curve, \
    average_precision_score

path2result = os.path.join("..", "results")
```

```python
def get_result(_result):
    print(f"{roc_auc_score(          _result['label'].values, _result['score'].values, average='macro'):.4f} -> rocauc")
    print(f"{average_precision_score(_result['label'].values, _result['score'].values, average='macro'):.4f} -> ap")

    print(f"{accuracy_score( _result['label'].values, (_result['score'] > 0.5).values):.4f} -> acc")
    print(f"{precision_score(_result['label'].values, (_result['score'] > 0.5).values):.4f} -> precision")
    print(f"{recall_score(   _result['label'].values, (_result['score'] > 0.5).values, zero_division=0):.4f} -> recall")
    print(f"{f1_score(       _result['label'].values, (_result['score'] > 0.5).values, zero_division=0):.4f} -> f1")
    
```

```python
metrics = ("AUC", "AP", "ACC", "Precision", "Recall", "F1")
penguin_means = {
    'GINEConv': (0.9894, 0.9832, 0.9585, 0.9712, 0.9006, 0.9346),
    'GENConv':  (0.9888, 0.9825, 0.9576, 0.9593, 0.9097, 0.9338),
    'GATConv':  (0.9284, 0.9291, 0.9446, 0.9982, 0.8330, 0.9081),
}
colors = ['salmon', 'wheat', 'thistle']
hatchs = ['//', '', '']

x = np.arange(len(metrics))  # the label locations
width = 0.13  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained', figsize=(5, 4))

i = 0
for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, color=colors[i], hatch=hatchs[i])
    # ax.bar_label(rects, padding=3)
    multiplier += 1
    i += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Metrics')
ax.set_title('Performance of different GNNs')
ax.set_xticks(x + width, metrics)
ax.legend()
ax.set_ylim(0.80, 1.)
ax.grid(True, linestyle=':', linewidth=.41)

plt.savefig('different_GNNs.eps', format='eps')
```

```python
import matplotlib.pyplot as plt  
import numpy as np  
import matplotlib.cm as cm  

# Data  
models = ['GATConv', 'GENConv', 'GINEConv']  
metrics = ['AUC', 'AP', 'Accuracy', 'Precision', 'Recall', 'F1']  
performance = [  
    [0.9273, 0.9277, 0.9445, 0.9981, 0.833, 0.9081],  # GATConv  
    [0.9878, 0.9811, 0.9551, 0.9343, 0.9289, 0.9316],  # GENConv  
    [0.9954, 0.9917, 0.9703, 0.9677, 0.941, 0.9542]    # GINEConv  
]  

# Plotting  
fig, ax = plt.subplots(figsize=(4, 3.5))  

# Set bar positions  
x = np.arange(len(metrics))  # the label locations  
width = 0.25  # the width of the bars  

# Define a color map
cmaps = [
    cm.get_cmap('Wistia'),
    cm.get_cmap('Oranges'),
    cm.get_cmap('Reds')
]
# cmap = cm.get_cmap('Blues')  

# Plot each model with a bar offset and color intensity based on value  
for i, model in enumerate(models):  
    colors = [cmaps[i](value) for value in performance[i]]  
    ax.bar(x + (i - 1) * width, performance[i], width, label=model, color=colors)  

# Customizations  
ax.set_xlabel('Metrics')  
ax.set_ylabel('Performance')  
ax.set_title('Performance of GNN Implementations')  
ax.set_xticks(x)  
ax.set_xticklabels(metrics, rotation=45, ha='right')  # Rotate labels  
ax.legend()  

# Add grid  
ax.yaxis.grid(True, linestyle='--', color='lightgrey', alpha=0.7)  

# Set y-axis limits  
min_value = min(min(performance))  
ax.set_ylim([min_value - 0.01, 1.0])  # Start slightly below the minimum value  

# Save as EPS with tight bounding box  
plt.tight_layout()  # this is to avoid clipping of ylabel  
plt.savefig('Performance_of_GNN_Implementations.eps', format='eps', bbox_inches='tight')  

# Display the plot  
plt.show()
```
